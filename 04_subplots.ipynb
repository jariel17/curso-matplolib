```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots()
axes.plot(x, y, 'c')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y, 'c')
axes[1].plot(y, x, 'r')
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
ax1.plot(x, y, 'c')
ax2.plot(y, x, 'r')
```

```python
fig, axes = plt.subplots(2, 4)
axes[0, 0].plot(x, np.cos(x), 'b')
axes[0, 1].plot(x, np.sin(x), 'r')
axes[0, 2].plot(x, np.tan(x), 'y')
axes[0, 3].plot(x, np.cos(x)**4, 'c')
```

```python
fig, ((ax1, ax2, ax3, ax4), (ax5, ax6, ax7, ax8)) = plt.subplots(2, 4)
ax1.plot(x, np.cos(x), 'b')
ax2.plot(x, np.sin(x), 'r')
ax3.plot(x, np.tan(x), 'y')
ax4.plot(x, np.cos(x)**4, 'c')
fig.tight_layout()
```

# Reto graficar la segunda linea

```python
fig, ((ax1, ax2, ax3, ax4), (ax5, ax6, ax7, ax8)) = plt.subplots(2, 4)
ax1.plot(x, np.cos(x), 'bd-')
ax2.plot(x, np.sin(x), 'rx-')
ax3.plot(x, np.tan(x), 'yo-')
ax4.plot(x, np.cos(x)**4, 'c')
ax5.plot(x, np.cos(x)*2, 'md:')
ax6.plot(x, np.sin(x)*2, 'o-')
ax7.plot(x, np.tan(x)*-1, 'salmon', linestyle=':')
ax8.plot(x, np.cos(x)*np.sin(x), 'co-')
fig.tight_layout()
```
