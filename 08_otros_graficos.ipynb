```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(0, 50, 100)
```

https://matplotlib.org/stable/gallery/index.html

## Histograma

```python
plt.hist(data, bins=10, histtype="step")
plt.show()
```

## Boxplot

```python
data = np.append(data, 200)
plt.boxplot(data, vert=False, patch_artist=True, notch=True, showfliers=True)
plt.show()
```

## Scatter

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N)) **2
colors = np.random.rand(N)
```

```python
plt.scatter(x, y, s=area, c=colors, marker='o', alpha=0.6)
plt.show()
```
