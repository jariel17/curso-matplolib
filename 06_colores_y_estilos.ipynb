```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
```

```python
plt.style.available
```

```python
plt.style.use('dark_background')
```

## Matlab

```python
fig, ax = plt.subplots(figsize=(8, 8))

ax.plot(x, x+1, 'r--')
ax.plot(x, x+2, 'b-')
ax.plot(x, x+3, 'y.-')
ax.plot(x, x+4, 'go:')
```

## Pyplot

```python
fig, ax = plt.subplots(figsize=(8, 8))

ax.plot(x, x+1, color='green',alpha=0.3, linewidth=8, linestyle='-', marker='o', markerfacecolor='white')
ax.plot(x, x+2, color='blue', linewidth=3, linestyle='dotted', marker='8')
ax.plot(x, x+3, color='#65FF34', linewidth=2, linestyle='dashdot', marker='v', markerfacecolor='red')
ax.plot(x, x+4, color='#21BBCC', linewidth=0.8, linestyle='dashed', marker='P', markersize=5)
```
