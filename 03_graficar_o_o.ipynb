```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x**2
```

```python
# La figura es como el lienzo en donde 'pintaremos' nuestras graficas
# axes son esas graficas
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.5, 0.5])
#first parameter: x position, second parameter: y position, third parameter x size, fourth parameter y size
axes.plot(x, y, 'b')
#fig.show() not required because we are on a Jupyter Notebook
```
