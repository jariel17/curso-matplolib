```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head(2)
```

```python
sns.countplot(data=tips, x='day', hue='sex')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.stripplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

```python
# Con variables numericas
sns.boxplot(data=tips, y='total_bill')
plt.show()
```

```python
# Con variables categoricas 
plt.figure(figsize=(6, 6))
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, color='black', marker='d')
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, palette='pastel', split=True)
plt.show()
```

```python
sns.violinplot(data=tips, y='total_bill')
```

catplot can be used to graph all the categories graphs using the parameter 'kind'  😸👇

```python
sns.catplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, palette='bright', kind='box', col='time')
plt.show()
```
